```python
# 5. Configuration and reproducibility
import os, random, numpy as np, torch
from src.config import Config

cfg = Config()
os.makedirs(cfg.output_dir, exist_ok=True)

random.seed(cfg.seed)
np.random.seed(cfg.seed)
torch.manual_seed(cfg.seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(cfg.seed)

# For initial debugging, keep these caps. For the complete selected 10%, set each to None.
print(cfg)
```

```python
# 6. Load and standardise datasets
from src.data import load_raid_10_percent, load_mage_10_percent

raid_all = load_raid_10_percent(
    fraction=cfg.raid_fraction,
    max_examples=cfg.max_raid_examples,
    seed=cfg.seed,
)
mage_retrieval, mage_external_test = load_mage_10_percent(
    fraction=cfg.mage_fraction,
    max_retrieval=cfg.max_mage_retrieval_examples,
    max_test=cfg.max_mage_test_examples,
    seed=cfg.seed,
)

print("RAID selected:", len(raid_all))
print("MAGE retrieval corpus:", len(mage_retrieval))
print("MAGE external test:", len(mage_external_test))
print("RAID label counts:", raid_all.to_pandas()["label"].value_counts().to_dict())
print("MAGE retrieval label counts:", mage_retrieval.to_pandas()["label"].value_counts().to_dict())
print("MAGE test label counts:", mage_external_test.to_pandas()["label"].value_counts().to_dict())

```

```python
# 7. Non-overlapping RAID train / validation / internal-test split
from datasets import ClassLabel

# Cast 'label' column to ClassLabel for stratification
raid_all = raid_all.cast_column("label", ClassLabel(num_classes=2, names=[ "human", "machine"]))

raid_train_temp = raid_all.train_test_split(
    test_size=0.20, seed=cfg.seed, stratify_by_column="label"
)
raid_train = raid_train_temp["train"]
raid_val_test = raid_train_temp["test"].train_test_split(
    test_size=0.50, seed=cfg.seed, stratify_by_column="label"
)
raid_val = raid_val_test["train"]
raid_test = raid_val_test["test"]

print({"train": len(raid_train), "validation": len(raid_val), "internal_test": len(raid_test)})
```

```python
# 8. Tokenisation
from transformers import AutoTokenizer, DataCollatorWithPadding

tokenizer = AutoTokenizer.from_pretrained(cfg.model_name)

def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        max_length=cfg.max_length,
    )

tokenized_train = raid_train.map(tokenize, batched=True, remove_columns=["text", "source", "domain"])
tokenized_val = raid_val.map(tokenize, batched=True, remove_columns=["text", "source", "domain"])
tokenized_test = raid_test.map(tokenize, batched=True, remove_columns=["text", "source", "domain"])
collator = DataCollatorWithPadding(tokenizer=tokenizer, pad_to_multiple_of=8)
```

```python
# 9. Fine-tune ModernBERT
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from src.modeling import classification_metrics

model = AutoModelForSequenceClassification.from_pretrained(
    cfg.model_name,
    num_labels=2,
    id2label={0: "human", 1: "machine"},
    label2id={"human": 0, "machine": 1},
)

args = TrainingArguments(
    output_dir=cfg.output_dir,
    learning_rate=cfg.learning_rate,
    per_device_train_batch_size=cfg.train_batch_size,
    per_device_eval_batch_size=cfg.eval_batch_size,
    gradient_accumulation_steps=cfg.gradient_accumulation_steps,
    num_train_epochs=cfg.num_train_epochs,
    weight_decay=cfg.weight_decay,
    warmup_ratio=cfg.warmup_ratio,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    greater_is_better=True,
    fp16=cfg.fp16 and torch.cuda.is_available(),
    logging_steps=50,
    save_total_limit=2,
    report_to="none",
    seed=cfg.seed,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    processing_class=tokenizer,
    data_collator=collator,
    compute_metrics=classification_metrics,
)
trainer.train()
trainer.save_model(cfg.output_dir + "/best_model")
tokenizer.save_pretrained(cfg.output_dir + "/best_model")
```

```python
# 10. Internal RAID evaluation
raid_metrics = trainer.evaluate(tokenized_test)
raid_metrics
```

```python
# 11. Baseline cross-dataset evaluation on MAGE (without retrieval)
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from src.modeling import predict_machine_probabilities

mage_texts = mage_external_test["text"]
mage_labels = np.asarray(mage_external_test["label"])

baseline_probs = predict_machine_probabilities(
    trainer.model, tokenizer, mage_texts,
    batch_size=cfg.eval_batch_size, max_length=cfg.max_length
)
baseline_preds = (baseline_probs >= 0.5).astype(int)

def full_metrics(y_true, y_pred):
    p, r, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="binary", zero_division=0
    )
    return {
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "precision": float(p),
        "recall": float(r),
        "f1": float(f1),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist(),
    }

baseline_metrics = full_metrics(mage_labels, baseline_preds)
baseline_metrics
```

```python
# 12. Build the MAGE retrieval index and query it
from src.modeling import embed_texts
from src.retrieval import LabelRetriever, blend_probabilities

model_path = cfg.output_dir + "/best_model"

retrieval_embeddings = embed_texts(
    model_path, mage_retrieval["text"],
    batch_size=cfg.eval_batch_size, max_length=cfg.max_length
)
query_embeddings = embed_texts(
    model_path, mage_texts,
    batch_size=cfg.eval_batch_size, max_length=cfg.max_length
)

retriever = LabelRetriever(k=cfg.retrieval_k).fit(
    retrieval_embeddings, mage_retrieval["label"]
)
retrieval_probs = retriever.predict_machine_probability(query_embeddings)
rag_probs = blend_probabilities(baseline_probs, retrieval_probs, alpha=cfg.rag_alpha)
rag_preds = (rag_probs >= 0.5).astype(int)

rag_metrics = full_metrics(mage_labels, rag_preds)
rag_metrics
```

```python
# 13. Compare baseline and retrieval-augmented results
import pandas as pd
results = pd.DataFrame([
    {"experiment": "RAID internal test", **{k:v for k,v in raid_metrics.items() if k.startswith("eval_")}},
    {"experiment": "MAGE external - ModernBERT only", **{k:v for k,v in baseline_metrics.items() if k != "confusion_matrix"}},
    {"experiment": "MAGE external - retrieval augmented", **{k:v for k,v in rag_metrics.items() if k != "confusion_matrix"}},
])
display(results)
```

```python
# 14. Confusion matrices
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

for title, preds in [
    ("MAGE: ModernBERT only", baseline_preds),
    ("MAGE: Retrieval-augmented", rag_preds),
]:
    fig, ax = plt.subplots(figsize=(5, 5))
    ConfusionMatrixDisplay.from_predictions(
        mage_labels, preds,
        display_labels=["human", "machine"],
        ax=ax,
        values_format="d",
    )
    ax.set_title(title)
    plt.show()
```
